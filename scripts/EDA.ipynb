Parte 2 — Análise de valores numéricos

Para a coluna "total", calcule:

- Valor mínimo
- Valor máximo
- Valor médio

```python
import pandas as pd
import duckdb

df = pd.read_csv('../datasets/vendas_2023_2024.csv')
df['sale_date'] = pd.to_datetime(df['sale_date'], dayfirst=True, errors='coerce')

query = """
SELECT
    COUNT(*)                                AS total_linhas,
    6                                       AS total_colunas,
    MIN(sale_date)                          AS data_minima,
    MAX(sale_date)                          AS data_maxima,
    MIN(total)                              AS total_minimo,
    MAX(total)                              AS total_maximo,
    ROUND(AVG(total), 2)                    AS total_medio,
    ROUND(MEDIAN(total), 2)                 AS total_mediana,
    ROUND(STDDEV(total), 2)                 AS total_desvio_padrao
FROM df
"""

duckdb.query(query).df()
```
