## Produtos

Tarefas: 

- Parte 1 — Padronize os nomes das categorias de produtos em: eletrônicos, propulsão e ancoragem.
- Parte 2 — Converta os valores para o tipo numérico.
- Parte 3 — Remova as duplicatas.

```python
import pandas as pd

#Visualizando os dados
df = pd.read_csv('../datasets/produtos_raw.csv')
df.tail()
```

```python
#Normalizando a categoria dos produtos
def normalizar_categoria(valor):
    # remove espaços internos, converte para minúsculo e tira acentos
    v = valor.replace(' ', '').lower()
    v = v.replace('ô', 'o').replace('ã', 'a').replace('ç', 'c')

    if 'eletr' in v:
        return 'eletrônicos'
    elif 'prop' in v:
        return 'propulsão'
    elif 'ancor' in v or 'encor' in v:
        return 'ancoragem'
    else:
        return 'desconhecido'

df['actual_category'] = df['actual_category'].apply(normalizar_categoria)
df['actual_category'].value_counts()
df.head()
```

```python
#Converta os valores para o tipo numérico
df['price'] = (
    df['price']
    .str.replace('R$', '', regex=False)
    .str.strip()
    .astype(float)
)

df.dtypes
```

```python
# Removendo linhas duplicadas
antes = len(df)
df = df.drop_duplicates()
depois = len(df)

print(f"Linhas removidas: {antes - depois}")
print(f"Linhas restantes: {depois}")
```
