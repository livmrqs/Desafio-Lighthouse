## Custos de Importação

Tarefa: 
- Carregue o arquivo JSON e gere um novo arquivo CSV organizando-o de acordo com a definição na imagem abaixo.

```python
# Carregando os dados 
import json
import pandas as pd

with open('../datasets/custos_importacao.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
# Transformando os dados em um DataFrame e organizando de acordo com o formato desejado
rows = []

for produto in data:
    for historico in produto['historic_data']:
        rows.append({
            'product_id':   produto['product_id'],
            'product_name': produto['product_name'],
            'category':     produto['category'],
            'start_date':   historico['start_date'],
            'usd_price':    historico['usd_price']
        })

df_custos = pd.DataFrame(rows)
df_custos.head(100)
```

```python
# Salvando o DataFrame em um arquivo CSV
df_custos.to_csv('../datasets/custos_importacao.csv', index=False)
print(f"Total de entradas: {len(df_custos)}")
```
